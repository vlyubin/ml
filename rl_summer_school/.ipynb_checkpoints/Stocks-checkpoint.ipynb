```python
# In this experiment we will be using a multi-armed bandit to buy stocks. At each day we choose a stock
# buy, based on the performance we've observed before, following the greedy-with-epsilon cutoff bandit strategy.
# We use the data for the last year (365 days) from quandl.

%matplotlib inline
import quandl
quandl.ApiConfig.api_key = "ZMRc36vKgtFiiH_aZk5b"
```

```python
stocks = ['AAPL', 'GOOG', 'MSFT', 'IBM', 'INTC', 'NVDA', 'ORCL', 'T', 'CSCO', 'FB']
```

```python
stock_data = {}
for s in stocks:
    stock_data[s] = quandl.get("WIKI/" + s, rows=365)
```

```python
import math
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

class StockArm():
    
  def __init__(self, symbol):
    self.symbol = symbol

  def step(self, day):
    return stock_data[self.symbol]['Close'][day] - stock_data[self.symbol]['Open'][day]

class StockBandit():

  def __init__(self, epsilon, k):
    self.epsilon = epsilon
    self.counts = [0] * k
    self.exp = [0.0] * k
    self.k = k

  def choose_action(self):
    if np.random.random() < self.epsilon:
      return np.random.randint(self.k)
    else:
      return np.argmax(self.exp)

  def update(self, idx, reward):
    self.exp[idx] = (self.exp[idx] * self.counts[idx] + reward) / (self.counts[idx] + 1)
    self.counts[idx] += 1

  def reset(self):
    self.counts = [0] * self.k
    self.exp = [0.0] * self.k

def test_bandit(bandit, arms, n_episodes, n_steps, *args, **kwargs):    
    chosen_arms = np.zeros((len(stocks), n_steps))
    avg_reward = np.zeros(n_steps)
    money_on_day = np.zeros(n_steps)

    for episode in range(n_episodes):
        ep_reward = 0
        cur_money = 1000.0
        bandit.reset()
        for step in range(n_steps):
            chosen_arm = bandit.choose_action()
            open_price = stock_data[arms[chosen_arm].symbol]['Open'][step]

            chosen_arms[chosen_arm][step] += 1
            reward = arms[chosen_arm].step(step)

            avg_reward[step] += reward * (cur_money / open_price)
            cur_money = cur_money + reward * (cur_money / open_price)
            money_on_day[step] += cur_money

            bandit.update(chosen_arm, reward)
    
    avg_reward /= n_episodes
    money_on_day /= n_episodes
    
    return chosen_arms, avg_reward, money_on_day
```

```python
arms = []
for i in range(len(stocks)):
    arms.append(StockArm(stocks[i]))

chosen_arms = {}
avg_reward = {}
money_on_day = {}

for e in [0.01, 0.05, 0.1, 0.2, 0.5, 1.0]:
    chosen_arms[e], avg_reward[e], money_on_day[e] = test_bandit(bandit = StockBandit(e, len(stocks)),
                                                        arms = arms, n_episodes = 10, n_steps = 365)
```

```python
import matplotlib.pyplot as plt

for e in money_on_day:
    plt.plot(money_on_day[e], label='e=' + str(e))

plt.legend(loc='best')
plt.ylabel('money_on_day')
plt.show()

# This plot shows how much money we would have by the last day for different values of e
```
